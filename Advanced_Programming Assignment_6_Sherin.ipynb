### Q1: You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more letter at a random position. Return the letter that was added to t.

```python
### Solution:

inp = input(" Please enter the first string : ")
inp1 = input(" Please enter the shuffled string : ")

def difference(a,b):
    """
    This function finds out the extra character that is there in the shuffled word
    """
    c = []
    d = []
    e = []
    for i in a:
        c.append(i)
    for j in b:
        d.append(j)
    for k in d:
        if d.count(k) > c.count(k):
            e.append(k) 
    yield set(e)

for g in difference(inp, inp1):
    print(g)
    

difference(inp,inp1) 
```

### Q2: Given a function that accepts unlimited arguments, check and count how many data types are in those arguments. Finally return the total in a list.

List order is:
[int, str, bool, list, tuple, dictionary]

```python
## Solution:

def count_datatype(*args):
    """
    This function counts the number of each data types and presents in a list
    """
    l=[]
    c = set()
    k=[]
    d2 = {"<class 'int'>" : 0, "<class 'str'>": 0, "<class 'bool'>": 0, "<class 'list'>":0, "<class 'tuple'>":0, "<class 'dict'>":0}
    for i in args:
        l.append(str(type(i)))
        c.add(str(type(i)))
    for j in c:
        if j in d2:
            d2[j]= l.count(j)
    k= list(d2.values())
    return k
```

### Q4: Given an integer between 0 and 26, make a variable (self.answer). That variable would be assigned to a string in this format:
"nines:your answer, threes:your answer, ones:your answer"
### You need to find out how many ones, threes, and nines it would at least take for the number of each to add up to the given integer when multiplied by one, three or nine (depends).

```python
## Solution:

try:
    inp = int(input(" Please enter an integer between 0 and 26 : "))
    def ones_thress_nines(x):
        d1 = {"nines":0, "threes":0, "ones":0}
        nines = int(x/9)
        threes = int((x - (nines*9))/3)
        ones = (x - (nines*9) - (threes*3))
        d1["nines"] = nines
        d1["threes"] = threes
        d1["ones"] = ones
        return d1
except ValueError:
    print(" Please enter a valid integer between 0 and 26")
    

ones_thress_nines(inp)
    
```

### Q5: The Fibonacci sequence is a classic use case for recursive functions since the value of the sequence at a given index is dependent on the sum of the last two values. However, the recursion tree created by solving the Fibonacci sequence recursively can grow quite fast. Therefore it can be important to think about the implications of running a function recursively. Depending on the size of n needed and the capabilities of the system in question you might want to take a different approach.

### Write a non-recursive function that takes an integer n and returns the Fibonacci sequence&#39;s value at index n.

```python
## Solution:

inp = int(input(" Please enter the integer position that you want to get the fibonacci number for : "))

def fib(x):
    
    a = 0
    b = 1
    for i in range(x+1):
        a, b = b, a+b
        yield a
z=[]       
for j in fib(inp):
    z.append(j)
print(z[inp-1]) 
    
        
```
